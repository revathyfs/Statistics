# **1. Overview of Data Types**

Data in statistics can be classified mainly into **two broad categories**:

1. **Numerical (Quantitative) Data:** Data represented by numbers and measurable quantities.
2. **Categorical (Qualitative) Data:** Data represented by categories or labels, not inherently numeric.

The type of data determines **which statistical methods and visualizations are appropriate**.

---

# **2. Numerical (Quantitative) Data**

Numerical data is measurable and allows **mathematical operations** like addition, subtraction, averaging, etc.

It is further divided into:

---

## **A. Discrete Data**

**Definition:** Discrete data can only take **specific, separate values**, often counts of items. It is **countable**.

**Characteristics:**

* Finite or countably infinite values.
* Often represents counts of objects or events.
* Cannot take fractional values (unless using rounded numbers for counting).

**Examples:**

| Scenario                          | Example Value |
| --------------------------------- | ------------- |
| Number of students in class       | 25, 30, 32    |
| Number of cars in a parking lot   | 10, 15, 20    |
| Number of goals scored in a match | 0, 1, 2, 3    |

**Visualization Techniques:**

* Bar charts, column charts, dot plots, or frequency tables.

**Implication in Analysis:**

* Mean, median, mode can be calculated.
* Cannot use continuous probability distributions (like normal) directly—often use **Poisson** or **Binomial distributions**.

---

## **B. Continuous Data**

**Definition:** Continuous data can take **any value within a range**, including fractions or decimals. It is **measurable**, not just countable.

**Characteristics:**

* Infinite possible values within a given range.
* Can be measured to any degree of precision.
* Often arises from measurements (height, weight, time).

**Examples:**

| Scenario                      | Example Value      |
| ----------------------------- | ------------------ |
| Height of students            | 150.2 cm, 172.5 cm |
| Temperature                   | 36.6°C, 37.2°C     |
| Time taken to complete a task | 12.5 min, 13.3 min |

**Visualization Techniques:**

* Histograms, box plots, line charts, scatter plots.

**Implication in Analysis:**

* Can use all statistical measures (mean, median, standard deviation).
* Often assumed to follow **normal distribution** for inferential statistics.

---

# **3. Categorical (Qualitative) Data**

Categorical data represents **qualities, attributes, or labels**, not numeric measurements. These cannot be meaningfully added or averaged.

It is divided into:

---

## **A. Nominal Data**

**Definition:** Nominal data represents **categories with no natural order or ranking**. Each category is just a label.

**Characteristics:**

* No inherent hierarchy or order.
* Categories are mutually exclusive.

**Examples:**

| Scenario   | Example Values     |
| ---------- | ------------------ |
| Eye color  | Blue, Brown, Green |
| Gender     | Male, Female       |
| Blood type | A, B, AB, O        |
| Country    | India, USA, France |

**Visualization Techniques:**

* Pie charts, bar charts, frequency tables.

**Implication in Analysis:**

* Can count occurrences in each category.
* Cannot calculate mean or median.
* Mode is meaningful (most frequent category).

---

## **B. Ordinal Data**

**Definition:** Ordinal data represents **categories with a meaningful order**, but the difference between categories is not necessarily equal or measurable.

**Characteristics:**

* Has a **rank/order**.
* Cannot meaningfully perform arithmetic operations.

**Examples:**

| Scenario              | Example Values                        |
| --------------------- | ------------------------------------- |
| Customer satisfaction | Poor, Average, Good, Excellent        |
| Education level       | High School < Bachelor < Master < PhD |
| Movie ratings         | ★, ★★, ★★★, ★★★★, ★★★★★               |

**Visualization Techniques:**

* Bar charts, stacked bar charts, heatmaps.
* Sometimes represented numerically for analysis (e.g., 1 = Poor, 2 = Average), but **interval assumptions are not guaranteed**.

**Implication in Analysis:**

* Median or percentiles can be used.
* Mean is **not recommended** unless the ordinal scale is treated as interval (with caution).

---

# **4. Quick Comparison Table**

| Feature    | Discrete (Numerical) | Continuous (Numerical) | Nominal (Categorical) | Ordinal (Categorical) |
| ---------- | -------------------- | ---------------------- | --------------------- | --------------------- |
| Nature     | Countable            | Measurable             | Label/Name            | Label with Order      |
| Values     | Integers only        | Any real number        | Categories            | Ordered categories    |
| Operations | +, -, mean, etc.     | +, -, mean, std, etc.  | Count/frequency       | Median, Percentile    |
| Example    | Number of cars       | Weight, Temperature    | Gender, Color         | Rating, Rank          |
| Chart Type | Bar chart, dot plot  | Histogram, line plot   | Pie chart, bar chart  | Bar chart, heatmap    |

---

**Key Takeaways:**

1. **Data type dictates analysis:** You can’t calculate mean of gender!
2. **Numerical vs categorical:** Numerical allows arithmetic; categorical does not.
3. **Discrete vs continuous:** Discrete is countable; continuous is measurable.
4. **Nominal vs ordinal:** Nominal has no order; ordinal has order but unequal spacing.

```python
# Continuous Data

import numpy as np

# Continuous data: heights of 10 people
heights = np.array([150.5, 160.2, 170.3, 165.0, 180.1, 155.5, 175.0, 168.2, 162.0, 177.5])
print("Heights:", heights)
print("Mean height:", np.mean(heights))
print("Standard deviation:", np.std(heights))
```

```python
# Discrete Data

# Discrete data: number of children in 10 families
children = np.array([2, 3, 1, 0, 4, 2, 3, 1, 2, 3])
print("Children:", children)
print("Mean:", np.mean(children))
print("Frequency count:", np.bincount(children))
```

```python
# Nominal Data

import pandas as pd

# Nominal data: favorite fruit
fruits = pd.Series(['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Apple'])
print("Fruits data:\n", fruits.value_counts())
```

```python
# Ordinal Data

from pandas.api.types import CategoricalDtype

ratings = pd.Series(['Poor', 'Good', 'Excellent', 'Good', 'Poor'])
rating_type = CategoricalDtype(categories=['Poor', 'Good', 'Excellent'], ordered=True)
ratings = ratings.astype(rating_type)
print("Ratings sorted:", ratings.sort_values())
```

### Summary and Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Numerical Data: Age and Sales
data = {
    'Favorite_Icecream': ['Vanilla', 'Chocolate', 'Strawberry', 'Vanilla'],
    'Calls_per_day': [10, 15, 12, 20],
    'Student_Height_cm': [160, 172, 168, 175],
    'Restaurant_Rating': ['★', '★★', '★★★', '★★★★'],
    'Monthly_Sales': [1000, 1500, 1200, 1800]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Visualizing categorical data
plt.figure(figsize=(6,4))
sns.countplot(x='Favorite_Icecream', data=df)
plt.title('Favorite Ice Cream Counts')
plt.show()
```

```python
# Visualizing numerical data
plt.figure(figsize=(6,4))
sns.histplot(df['Student_Height_cm'], bins=5, kde=True)
plt.title('Student Height Distribution')
plt.show()
```

### Mixed Data Analysis

```python
# Mixed data example
data = {
    'Employee': ['A', 'B', 'C', 'D', 'E'],
    'Age': [25, 30, 28, 35, 40],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'Department': ['HR', 'IT', 'IT', 'Sales', 'HR'],
    'Experience_years': [2, 5, 3, 10, 12],
    'Performance_Rating': ['Excellent', 'Good', 'Excellent', 'Average', 'Good']
}
```

```python
df = pd.DataFrame(data)
```

```python
# Identify data types
print(df.dtypes)
```

```python
# Summary statistics for numerical data
print("\nNumerical Summary:")
print(df[['Age', 'Experience_years']].describe())
```

```python
# Countplot for categorical data
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Performance_Rating', data=df, order=['Average','Good','Excellent'])
plt.title('Performance Rating Distribution')
plt.show()
```

```python
# Boxplot to compare numerical across categories
plt.figure(figsize=(6,4))
sns.boxplot(x='Department', y='Experience_years', data=df)
plt.title('Experience by Department')
plt.show()
```

### Real-Life Online Store Example

```python
# Import libraries
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Numerical Data
ages = np.array([23, 25, 31, 28, 22, 24, 30, 29, 27, 26])
print("Mean Age:", np.mean(ages))
print("Median Age:", np.median(ages))
print("Std Deviation:", np.std(ages))
```

```python
# Categorical Data
colors = pd.Series(['Red','Blue','Green','Red','Blue','Red'])
print("Color counts:\n", colors.value_counts())
```

```python
# Ordinal Data
ratings = pd.Series(['Poor','Good','Excellent','Good','Poor'])
rating_type = pd.CategoricalDtype(categories=['Poor','Good','Excellent'], ordered=True)
ratings = ratings.astype(rating_type)
print("Sorted Ratings:\n", ratings.sort_values())
```

```python
# Exercise: Histogram for continuous data
temps = np.random.normal(37, 1, 100)  # 100 temperature readings
plt.hist(temps, bins=10, color='skyblue', edgecolor='black')
plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```

```python
mode_result = stats.mode(temps, keepdims=True)
print("Mode of temperature:", mode_result.mode[0])
print("Count of mode:", mode_result.count[0])
```
