```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

----------------------------
# Section 1: Central Tendency
----------------------------

```python
# Exercise 1: Calculate mean, median, mode of exam scores
scores = np.array([45, 50, 55, 60, 60, 65, 70, 75, 80, 90, 95])
print("Exercise 1 - Mean:", np.mean(scores))
print("Median:", np.median(scores))
print("Mode:", stats.mode(scores, keepdims=True).mode[0])
```

```python
# Exercise 2: Create dataset of salaries and calculate central tendency
salaries = np.random.randint(30000, 120000, 15)
print("\nExercise 2 - Salaries:", salaries)
print("Mean:", np.mean(salaries))
print("Median:", np.median(salaries))
print("Mode:", stats.mode(salaries, keepdims=True).mode[0])
```

```python
# Exercise 3: Find median house price from dataset
house_prices = np.array([120000, 150000, 200000, 250000, 300000, 1000000])
print("\nExercise 3 - Median House Price:", np.median(house_prices))
```

```python
# Exercise 4: Compare mean vs median in skewed dataset
skewed = np.array([5, 6, 7, 8, 100])
print("\nExercise 4 - Mean:", np.mean(skewed), "Median:", np.median(skewed))
```

```python
# Exercise 5: Find mode of favorite fruit list
fruits = pd.Series(['Apple','Banana','Apple','Orange','Banana','Apple','Mango'])
print("\nExercise 5 - Fruit Mode:", fruits.mode()[0])
```

----------------------------
# Section 2: Dispersion
----------------------------

```python
# Exercise 6: Calculate range of exam scores
print("\nExercise 6 - Range:", np.max(scores)-np.min(scores))
```

```python
# Exercise 7: Variance and standard deviation of salaries
print("\nExercise 7 - Variance:", np.var(salaries))
print("Standard Deviation:", np.std(salaries))
```

```python
# Exercise 8: Find IQR of exam scores
Q1, Q3 = np.percentile(scores, [25, 75])
print("\nExercise 8 - IQR:", Q3-Q1)
```

```python
# Exercise 9: Compare spread of two datasets
set1 = np.random.randint(50,60,10)
set2 = np.random.randint(10,100,10)
print("\nExercise 9 - Std Dev set1:", np.std(set1))
print("Std Dev set2:", np.std(set2))
```

```python
# Exercise 10: Detect outliers using IQR method
Q1, Q3 = np.percentile(house_prices, [25, 75])
IQR = Q3-Q1
outliers = house_prices[(house_prices < Q1 - 1.5*IQR) | (house_prices > Q3 + 1.5*IQR)]
print("\nExercise 10 - Outliers:", outliers)
```

----------------------------
# Section 3: Visualization
----------------------------

```python
# Exercise 11: Plot histogram of exam scores
plt.hist(scores, bins=5, color='lightblue', edgecolor='black')
plt.title("Exercise 11 - Histogram of Scores")
plt.show()
```

```python
# Exercise 12: Plot boxplot of salaries
plt.boxplot(salaries)
plt.title("Exercise 12 - Boxplot of Salaries")
plt.show()
```

```python
# Exercise 13: Pie chart of fruit counts
fruits.value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Exercise 13 - Fruit Pie Chart")
plt.show()
```

```python
# Exercise 14: Bar chart of fruit counts
fruits.value_counts().plot.bar(color='orange')
plt.title("Exercise 14 - Fruit Bar Chart")
plt.show()
```

```python
# Exercise 15: Compare boxplots of two datasets
plt.boxplot([set1, set2], labels=['Set1','Set2'])
plt.title("Exercise 15 - Compare Boxplots")
plt.show()
```

```python
# Exercise 19: Group salaries into bins and count
salary_bins = pd.cut(salaries, bins=[0,50000,80000,120000], labels=['Low','Medium','High'])
print("\nExercise 19 - Salary Groups:\n", salary_bins.value_counts())
```

```python
# Exercise 20: Plot bar chart of salary groups
salary_bins.value_counts().plot.bar(color='green')
plt.title("Exercise 20 - Salary Groups Bar Chart")
plt.show()
```
