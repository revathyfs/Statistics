```python
# Exercise 1

# A coin is tossed once.

# P(Head) = 0.5, P(Tail) = 0.5

import random

# Simulation
trials = 100000
heads = sum(1 for _ in range(trials) if random.choice(["H","T"])=="H")
print("P(Head) ~", heads/trials)
print("P(Tail) ~", 1 - heads/trials)
```

```python
# Exercise 2

# Die rolled once.

trials = 100000
greater4 = sum(1 for _ in range(trials) if random.randint(1,6)>4)
odd = sum(1 for _ in range(trials) if random.randint(1,6)%2!=0)

print("P(number > 4) ~", greater4/trials)
print("P(odd) ~", odd/trials)
```

```python
# Exercise 3

# Bag = {3R, 2B, 5G}

colors = ["R"]*3 + ["B"]*2 + ["G"]*5
trials = 100000
draws = [random.choice(colors) for _ in range(trials)]

print("P(Red) ~", draws.count("R")/trials)
print("P(Green) ~", draws.count("G")/trials)
print("P(Red or Blue) ~", (draws.count("R")+draws.count("B"))/trials)
```

```python
# Exercise 4

# Deck of 52 cards

import itertools

deck = [(s,r) for s in ['H','D','C','S'] for r in range(1,14)]
trials = 100000
draws = [random.choice(deck) for _ in range(trials)]

aces = sum(1 for c in draws if c[1]==1)
hearts = sum(1 for c in draws if c[0]=='H')
aces_kings = sum(1 for c in draws if c[1] in [1,13])

print("P(Ace) ~", aces/trials)
print("P(Heart) ~", hearts/trials)
print("P(Ace or King) ~", aces_kings/trials)
```

```python
# Exercise 5

# Toss 2 coins.

trials = 100000
both_heads = sum(1 for _ in range(trials) if random.choice(["H","T"])=="H" and random.choice(["H","T"])=="H")
at_least1 = sum(1 for _ in range(trials) if "H" in [random.choice(["H","T"]), random.choice(["H","T"])])

print("P(Both heads) ~", both_heads/trials)
print("P(At least one head) ~", at_least1/trials)
```

```python
# Exercise 6

# Two dice rolled.

trials = 100000
both_even = sum(1 for _ in range(trials) if random.randint(1,6)%2==0 and random.randint(1,6)%2==0)
at_least6 = sum(1 for _ in range(trials) if 6 in [random.randint(1,6), random.randint(1,6)])

print("P(Both even) ~", both_even/trials)
print("P(At least one 6) ~", at_least6/trials)
```

```python
# Exercise 8

# Two dice, sum=7.

outcomes = [(i,j) for i in range(1,7) for j in range(1,7)]
favorable = [(i,j) for (i,j) in outcomes if i+j==7]
print("Theoretical P(sum=7):", len(favorable)/36)

```

```python
# Exercise 11

# Two balls drawn (without replacement).

balls = ["R"]*6 + ["B"]*4
trials = 100000
both_red = 0
red_blue = 0

for _ in range(trials):
    draw = random.sample(balls,2)
    if draw[0]=="R" and draw[1]=="R":
        both_red+=1
    if draw[0]=="R" and draw[1]=="B":
        red_blue+=1

print("P(Both Red) ~", both_red/trials)
print("P(Red then Blue) ~", red_blue/trials)
```

```python
# Bayesâ€™ Theorem

# Machine A: 60%, defective=2%
# Machine B: 40%, defective=5%

P_A = 0.6
P_B = 0.4
P_D_A = 0.02
P_D_B = 0.05

P_def = P_A*P_D_A + P_B*P_D_B
P_B_given_D = (P_B*P_D_B)/P_def
print("P(Machine B | defective):", P_B_given_D)
```

```python
# 2 Dice, sum=7 & 11

trials = 20000
sum7 = 0
sum11 = 0

for _ in range(trials):
    total = random.randint(1,6) + random.randint(1,6)
    if total==7: sum7+=1
    if total==11: sum11+=1

print("P(sum=7) ~", sum7/trials)
print("P(sum=11) ~", sum11/trials)
```

```python
# Monty Hall Simulatio

import numpy as np

def monty(trials=10000):
    switch_wins = 0
    stay_wins = 0
    for _ in range(trials):
        doors = [0,0,1]  # 1 = car
        np.random.shuffle(doors)
        choice = np.random.randint(0,3)
        # host opens a goat door
        remaining = [i for i in range(3) if i!=choice and doors[i]==0]
        host = np.random.choice(remaining)
        # switch choice
        switch = [i for i in range(3) if i not in [choice, host]][0]
        if doors[switch]==1: switch_wins+=1
        if doors[choice]==1: stay_wins+=1
    return switch_wins/trials, stay_wins/trials

print("Switch win probability:", monty()[0])
print("Stay win probability:", monty()[1])
```
