```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

----------------------------
# Section 1: Numerical Data (Continuous & Discrete)
----------------------------

```python
# Exercise 1: Create an array of ages and calculate mean, median, std
a_ages = np.array([23, 25, 31, 28, 22, 24, 30, 29, 27, 26])
```

```python
print("Exercise 1 - Ages:", a_ages)
```

```python
print("Mean:", np.mean(a_ages))
```

```python
print("Median:", np.median(a_ages))
```

```python
print("Std Dev:", np.std(a_ages))
```

```python
# Exercise 2: Generate 15 random weights and find min, max, range
weights = np.random.uniform(50, 90, 15)
```

```python
print("\nExercise 2 - Weights:", weights)
```

```python
print("Min:", np.min(weights), "Max:", np.max(weights), "Range:", np.max(weights)-np.min(weights))
```

```python
# Exercise 3: Discrete data - Number of children in families
children = np.array([2, 3, 1, 0, 4, 2, 3, 1, 2, 3])
print("\nExercise 3 - Children frequency:", np.bincount(children))
```

```python
# Exercise 4: Plot histogram of 50 exam scores
scores = np.random.randint(40, 100, 50)
plt.hist(scores, bins=10, color='lightgreen', edgecolor='black')
plt.title("Exercise 4 - Exam Score Distribution")
plt.show()
```

```python
# Exercise 5: Find mean, mode of exam scores
mode_scores = stats.mode(scores, keepdims=True)
print("\nExercise 5 - Mean:", np.mean(scores))
print("Mode:", mode_scores.mode[0], "Count:", mode_scores.count[0])
```

----------------------------
# Section 2: Categorical Data (Nominal & Ordinal)
----------------------------

```python
# Exercise 6: Nominal data - Favorite colors
colors = pd.Series(['Red','Blue','Green','Red','Blue','Red','Green','Green'])
print("\nExercise 6 - Color counts:\n", colors.value_counts())
```

```python
# Exercise 7: Nominal data - Count most common fruit
fruits = pd.Series(['Apple','Banana','Apple','Orange','Banana','Apple','Mango'])
print("\nExercise 7 - Fruit counts:\n", fruits.value_counts())
```

```python
# Exercise 8: Ordinal data - Ratings
from pandas.api.types import CategoricalDtype
ratings = pd.Series(['Poor','Good','Excellent','Good','Poor','Excellent','Good'])
rating_type = CategoricalDtype(categories=['Poor','Good','Excellent'], ordered=True)
ratings = ratings.astype(rating_type)
print("\nExercise 8 - Sorted ratings:\n", ratings.sort_values())
```

```python
# Exercise 9: Education level (Ordinal)
edu = pd.Series(['Bachelor','High School','Master','PhD','High School','Master'])
edu_type = CategoricalDtype(categories=['High School','Bachelor','Master','PhD'], ordered=True)
edu = edu.astype(edu_type)
print("\nExercise 9 - Education sorted:\n", edu.sort_values())
```

# Section 3: Mixed Dataset Practice
----------------------------

```python
# Exercise 10: Create student dataset with Age, Gender, Height, Grade
students = pd.DataFrame({
'Age': np.random.randint(18, 25, 10),
'Gender': np.random.choice(['Male','Female'], 10),
'Height': np.random.normal(165, 10, 10),
'Grade': np.random.choice(['A','B','C','D'], 10)
})
print("\nExercise 10 - Student dataset:\n", students)
```

```python
# Exercise 11: Average height by gender
print("\nExercise 11 - Avg height by gender:\n", students.groupby('Gender')['Height'].mean())
```

```python
# Exercise 12: Count frequency of grades
print("\nExercise 12 - Grade counts:\n", students['Grade'].value_counts())
```

```python
# Exercise 13: Sort students by age and grade
print("\nExercise 13 - Sorted students:\n", students.sort_values(['Age','Grade']))
```

```python
# Exercise 14: Generate 100 random continuous temps, plot histogram
temps = np.random.normal(37, 1, 100)
plt.hist(temps, bins=10, color='skyblue', edgecolor='black')
plt.title("Exercise 14 - Temperature Distribution")
plt.show()
```

```python
# Exercise 15: Find mean, median, mode of temps
mode_temp = stats.mode(temps, keepdims=True)
print("\nExercise 15 - Mean:", np.mean(temps))
print("Median:", np.median(temps))
print("Mode:", mode_temp.mode[0])
```

```python
# Exercise 29: Generate dataset of 50 people with categorical + numeric fields
people = pd.DataFrame({
'Age': np.random.randint(18,60,50),
'Department': np.random.choice(['HR','IT','Finance','Sales'],50),
'Salary': np.random.randint(30000,100000,50)
})
print("\nExercise 29 - People dataset:\n", people.head())
```

```python
# Exercise 30: Group average salary by Department
print("\nExercise 30 - Avg Salary by Department:\n", people.groupby('Department')['Salary'].mean())
```

```python
# Exercise 31: Create 40 random categorical colors and plot frequency
colors2 = pd.Series(np.random.choice(['Red','Blue','Green','Yellow'],40))
colors2.value_counts().plot(kind='bar')
plt.title("Exercise 31 - Color Frequency")
plt.show()
```

```python
# Exercise 32: Standardize numerical data (Age, Salary)
people[['Age_z','Salary_z']] = (people[['Age','Salary']] - people[['Age','Salary']].mean()) / people[['Age','Salary']].std()
print("\nExercise 32 - Standardized values:\n", people.head())
```
