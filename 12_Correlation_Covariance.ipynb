
# **1ï¸âƒ£ Covariance**

### **Definition**

Covariance measures how **two variables change together**. If one variable tends to increase when the other increases, the covariance is positive. If one tends to decrease when the other increases, the covariance is negative.

**Formula (for population):**
$[
\text{Cov}(X, Y) = \frac{\sum_{i=1}^{N} (X_i - \mu_X)(Y_i - \mu_Y)}{N}
]$

**Sample version:**
$[
\text{Cov}(X, Y) = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{n-1}
]$

Where:

* $(X_i, Y_i)$ are data points
* $(\mu_X, \mu_Y)$ or $(\bar{X}, \bar{Y})$ are means of X and Y

### **Interpretation**

* **Cov > 0** â†’ Variables move in the **same direction**.
* **Cov < 0** â†’ Variables move in **opposite directions**.
* **Cov â‰ˆ 0** â†’ Variables are **uncorrelated** (no linear relationship).

ðŸ’¡ **Problem with covariance:** Its magnitude depends on the units of X and Y, making it hard to interpret directly.

# **2ï¸âƒ£ Correlation**

Correlation is a **standardized version of covariance** that measures both **strength** and **direction** of a linear relationship between variables.

### **2a: Pearson Correlation**

**Formula:**
$[
r = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
]$

Where $(\sigma_X)$ and $(\sigma_Y)$ are standard deviations of X and Y.

* Range: ([-1, 1])

  * (r = 1) â†’ Perfect positive linear relationship
  * (r = -1) â†’ Perfect negative linear relationship
  * (r = 0) â†’ No linear relationship

**Intuition:**

* Measures **how well a straight line fits the data**.
* Sensitive to **outliers** and assumes **linear relationship**.

---

### **Example**

| X | 1 | 2 | 3 | 4 | 5  |
| - | - | - | - | - | -- |
| Y | 2 | 4 | 6 | 8 | 10 |

* Cov(X,Y) = 5
* $(Ïƒ_X)$ = 1.58, $(Ïƒ_Y)$ = 3.16
* Pearson r = 5 / (1.58 Ã— 3.16) â‰ˆ 1 â†’ Perfect positive correlation

---

### **2b: Spearman Rank Correlation**

Pearson measures **linear relationships**. But sometimes, relationships are **monotonic but not linear** (e.g., exponential growth). Spearman correlation handles this by using **ranks** instead of actual values.

**Steps to calculate Spearman correlation:**

1. Replace each data point with its **rank**.
2. Compute **Pearson correlation** of the ranks.

**Formula (simplified if no ties):**
$[
\rho = 1 - \frac{6 \sum d_i^2}{n(n^2-1)}
]$
Where $(d_i = \text{rank}(X_i) - \text{rank}(Y_i))$

* Range: ([-1, 1])
* Ï = 1 â†’ Perfect monotonic increase
* Ï = -1 â†’ Perfect monotonic decrease

ðŸ’¡ **Use case:** Spearman is **robust to outliers** and detects **monotonic relationships**, not just linear ones.

---

### **Example**

| X | 1 | 2 | 3 | 4  | 5  |
| - | - | - | - | -- | -- |
| Y | 1 | 4 | 9 | 16 | 25 |

* Clearly **non-linear** but **monotonic**.
* Pearson r â‰ˆ 0.97 (not perfect linear)
* Spearman Ï = 1 (perfect monotonic)

# **3ï¸âƒ£ Key Differences: Pearson vs Spearman**

| Feature                 | Pearson                                           | Spearman                                             |
| ----------------------- | ------------------------------------------------- | ---------------------------------------------------- |
| Relationship detected   | Linear                                            | Monotonic (linear or non-linear)                     |
| Sensitivity to outliers | High                                              | Low                                                  |
| Data requirement        | Interval/ratio scale                              | Ordinal, interval, or ratio                          |
| Range                   | -1 to 1                                           | -1 to 1                                              |
| Interpretation          | Strength and direction of **linear** relationship | Strength and direction of **monotonic** relationship |

# **4ï¸âƒ£ How to Understand Variable Relationships**

1. **Scatter plot** â†’ First, visualize relationship

   * Linear trend â†’ Pearson
   * Non-linear monotonic â†’ Spearman
2. **Compute correlation coefficient** â†’ Quantify strength and direction
3. **Covariance** â†’ Understand raw co-movement (but not standardized)
4. **Check assumptions** â†’ Pearson assumes linearity; Spearman does not

# **5ï¸âƒ£ Python Implementation**

```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, spearmanr
```

```python
# Sample data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 6, 8, 10])
Z = np.array([1, 4, 9, 16, 25])
```

```python
# Covariance
cov_XY = np.cov(X, Y)[0, 1]
print("Covariance X,Y:", cov_XY)
```

```python
# Pearson correlation
pearson_XY = pearsonr(X, Y)[0]
print("Pearson correlation X,Y:", pearson_XY)
```

```python
# Spearman correlation
spearman_XZ = spearmanr(X, Z)[0]
print("Spearman correlation X,Z:", spearman_XZ)
```
