# ğŸ“Œ 1. Inferential Statistics Basics

Inferential statistics = making conclusions about a **population** using a **sample**.
Key tools:

1. **Hypothesis testing** â†’ decision-making.
2. **Confidence intervals** â†’ estimation with uncertainty.

---

## ğŸ”¹ Hypothesis Testing

A structured procedure to test assumptions about population parameters.

### General steps

1. **State hypotheses**:

   * Null hypothesis $( H_0 )$: â€œno effect / no difference.â€
   * Alternative hypothesis $( H_1 )$: â€œthere is an effect / difference.â€

2. **Choose significance level $((\alpha))$**: often 0.05.

3. **Select test statistic**: t, F, Chi-square, etc.

4. **Compute test statistic & p-value**.

5. **Decision**:

   * If p < Î± â†’ Reject $(H_0)$.
   * If p â‰¥ Î± â†’ Fail to reject $(H_0)$.

---

### ğŸ§ª t-test

Used when comparing **means**.

* **One-sample t-test**: compare sample mean vs population mean.
* **Independent two-sample t-test**: compare means of two independent groups.
* **Paired t-test**: compare means of the same group at two times.

Formula (one-sample):
$[
t = \frac{\bar{x} - \mu_0}{s/\sqrt{n}}
]$

---

### ğŸ§ª Chi-square test

Used for **categorical data**.

* **Goodness of fit**: does observed frequency match expected?
* **Test of independence**: are two categorical variables independent?

Formula:
$[
\chi^2 = \sum \frac{(O - E)^2}{E}
]$
where O = observed, E = expected frequencies.

---

### ğŸ§ª ANOVA (Analysis of Variance)

Compares **means of more than 2 groups**.

* Null hypothesis: all group means equal.
* Alternative: at least one mean is different.

Statistic:
$[
F = \frac{\text{Between-group variability}}{\text{Within-group variability}}
]$

If F is large â†’ evidence against $(H_0)$.

---

## ğŸ”¹ Confidence Intervals (CI)

A **range of values** within which the true population parameter is likely to fall.

Example: 95% CI for population mean:
$[
\bar{x} \pm z_{\alpha/2}\frac{s}{\sqrt{n}}
]$

Interpretation: If we repeatedly sample, ~95% of such intervals would contain the true mean.

### Example 1 â€” One-sample t-test

```python
import numpy as np
from scipy import stats

# Suppose we want to test if average weight = 70 kg
data = np.random.normal(72, 5, 30)  # sample of 30 students

t_stat, p_val = stats.ttest_1samp(data, 70)
print("t-statistic:", t_stat, "p-value:", p_val)
```

### Example 2 â€” Independent two-sample t-test

```python
group1 = np.random.normal(70, 5, 30)  # sample 1
group2 = np.random.normal(73, 5, 30)  # sample 2

t_stat, p_val = stats.ttest_ind(group1, group2)
print("t-statistic:", t_stat, "p-value:", p_val)
```

### Example 3 â€” Paired t-test

```python
before = np.random.normal(60, 5, 20)
after = before + np.random.normal(2, 3, 20)  # improvement by ~2

t_stat, p_val = stats.ttest_rel(before, after)
print("Paired t-test:", t_stat, p_val)
```

### Example 4 â€” Chi-square test of independence

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Example: Gender vs Preference
data = [[30, 10],   # Male: like/dislike
        [25, 15]]   # Female: like/dislike
df = pd.DataFrame(data, columns=["Like", "Dislike"], index=["Male","Female"])

chi2, p, dof, expected = chi2_contingency(df)
print("Chi2 statistic:", chi2, "p-value:", p)
```

### Example 5 â€” One-way ANOVA

```python
groupA = np.random.normal(50, 5, 20)
groupB = np.random.normal(52, 5, 20)
groupC = np.random.normal(55, 5, 20)

f_stat, p_val = stats.f_oneway(groupA, groupB, groupC)
print("F-statistic:", f_stat, "p-value:", p_val)
```

### Example 6 â€” Confidence Interval for a mean

```python
sample = np.random.normal(100, 15, 40)
xbar = np.mean(sample)
s = np.std(sample, ddof=1)
n = len(sample)
alpha = 0.05

# 95% CI using t-distribution
t_val = stats.t.ppf(1 - alpha/2, df=n-1)
ci_lower = xbar - t_val * s/np.sqrt(n)
ci_upper = xbar + t_val * s/np.sqrt(n)
print("95% Confidence Interval:", (ci_lower, ci_upper))
```

### Generate two groups of exam scores and test if their means differ using t-test.


```python
import numpy as np
from scipy import stats

# Generate two groups
np.random.seed(0)
group1 = np.random.normal(70, 10, 30)   # mean 70
group2 = np.random.normal(75, 12, 30)   # mean 75

t_stat, p_val = stats.ttest_ind(group1, group2)
print("t-statistic:", round(t_stat, 3), "p-value:", round(p_val, 4))
```

### Run a Chi-square test on a 3Ã—2 contingency table (e.g., 3 education levels Ã— pass/fail).

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Education level Ã— Pass/Fail
data = [[40, 10],  # High school
        [50, 20],  # Undergraduate
        [30, 25]]  # Postgraduate

df = pd.DataFrame(data, columns=["Pass", "Fail"],
                  index=["HighSchool", "Undergrad", "Postgrad"])

chi2, p, dof, expected = chi2_contingency(df)
print("Chi2:", round(chi2, 3), "p-value:", round(p, 4))
print("Expected frequencies:\n", expected)
```

### Perform ANOVA on 4 groups and check if means differ.

```python
np.random.seed(1)
groupA = np.random.normal(50, 5, 20)
groupB = np.random.normal(52, 5, 20)
groupC = np.random.normal(55, 5, 20)
groupD = np.random.normal(60, 5, 20)

f_stat, p_val = stats.f_oneway(groupA, groupB, groupC, groupD)
print("F-statistic:", round(f_stat, 3), "p-value:", round(p_val, 4))
```

### Build a function to compute confidence interval for any sample mean.

```python
def mean_confidence_interval(data, confidence=0.95):
    import numpy as np
    from scipy import stats
    
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    se = stats.sem(data)  # standard error
    h = se * stats.t.ppf((1 + confidence) / 2., n-1)
    return mean, mean-h, mean+h

# Example usage
sample = np.random.normal(100, 15, 40)
mean, lower, upper = mean_confidence_interval(sample)
print(f"Sample mean = {mean:.2f}, 95% CI = ({lower:.2f}, {upper:.2f})")
```
