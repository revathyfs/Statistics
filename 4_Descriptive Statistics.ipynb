

# üìä Descriptive Statistics

Descriptive statistics summarizes and describes the **main features** of a dataset.
It answers **‚ÄúWhat does my data look like?‚Äù** before moving to advanced modeling.

We divide it into three main parts:

---

## **1Ô∏è‚É£ Central Tendency (the "middle" of data)**

Central tendency gives a **single representative value** of the dataset.

### **(a) Mean (Average)**

* Formula:

  $[
  \text{Mean} = \frac{\sum_{i=1}^{n} x_i}{n}
  ]$

* Meaning: Add all values, divide by count.
* Sensitive to **outliers** (e.g., salaries: [30k, 32k, 35k, 1M] ‚Üí mean is misleading).

‚úÖ Example:
Data = [10, 20, 30]
Mean = (10+20+30)/3 = **20**

---

### **(b) Median (Middle value)**

* Arrange data in order ‚Üí pick middle.
* If even count ‚Üí average of 2 middle values.
* **Robust to outliers**.

‚úÖ Example:
Data = [10, 20, 100] ‚Üí Median = 20 (better than mean = 43.3)

---

### **(c) Mode (Most frequent value)**

* The value occurring **most often**.
* Useful for **categorical data** (e.g., most common color, product).
* A dataset may have:

  * **One mode** (unimodal)
  * **Multiple modes** (bimodal/multimodal)
  * **No mode** (if all frequencies equal)

‚úÖ Example:
Data = [10, 10, 20, 30] ‚Üí Mode = 10

---

## **2Ô∏è‚É£ Dispersion (spread of data)**

Central tendency alone is not enough ‚Äî two datasets with same mean can have **different spread**.
Dispersion shows **how data varies**.

---

### **(a) Range**

* Formula:
  
  $[
  \text{Range} = \text{Max} - \text{Min}
  ]$

* Simple but sensitive to outliers.

‚úÖ Example: [10, 12, 15, 18] ‚Üí Range = 18-10 = 8

---

### **(b) Variance**

* Formula:
  
  $[
  \sigma^2 = \frac{\sum (x_i - \mu)^2}{n}
  ]$

* Measures **average squared deviation** from mean.
* Units are **squared** (e.g., cm¬≤).

---

### **(c) Standard Deviation (SD)**

* Formula:
  
  $[
  \sigma = \sqrt{\text{Variance}}
  ]$

* Easier to interpret because it‚Äôs in **same units** as data.
* Low SD ‚Üí values close to mean.
* High SD ‚Üí values spread out.

---

### **(d) Interquartile Range (IQR)**

* Focuses on **middle 50% of data**.
* Formula:
  
  $[
  \text{IQR} = Q3 - Q1
  ]$
  
  where

  * Q1 = 25th percentile (lower quartile)
  * Q3 = 75th percentile (upper quartile)
* Removes outlier influence.

‚úÖ Example: [1, 2, 3, 4, 5, 6, 7, 8, 9]

* Q1 = 3, Q3 = 7 ‚Üí IQR = 4

---

## **3Ô∏è‚É£ Visualization**

Visuals help **understand distribution & spread** quickly.

### **(a) Histogram**

* Shows frequency distribution of continuous data.
* Bars represent intervals (bins).
* Useful to check **skewness, spread, modality**.

### **(b) Boxplot (Whisker plot)**

* Summarizes data using **median, quartiles, min, max, outliers**.
* Good for detecting **outliers** and comparing groups.

### **(c) Pie Chart**

* Good for **categorical proportions**.
* Limited use in professional analytics (bar charts often better).

### **(d) Bar Chart**

* For **categorical data** comparison.
* Length of bars shows frequency/count.

---

## ‚úÖ Python Example

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Sample data: exam scores
scores = np.array([45, 50, 55, 60, 60, 65, 70, 75, 80, 90, 95])
```

```python
# Central Tendency
print("Mean:", np.mean(scores))
print("Median:", np.median(scores))
print("Mode:", stats.mode(scores, keepdims=True).mode[0])
```

```python
# Dispersion
print("Range:", np.max(scores)-np.min(scores))
print("Variance:", np.var(scores))
print("Standard Deviation:", np.std(scores))
Q1, Q3 = np.percentile(scores, [25, 75])
print("IQR:", Q3-Q1)
```

```python
# Visualization
plt.figure(figsize=(12,4))

# Histogram
plt.subplot(1,3,1)
plt.hist(scores, bins=5, color='lightblue', edgecolor='black')
plt.title("Histogram")
```

```python
# Visualization
plt.figure(figsize=(12,4))

# Boxplot
plt.subplot(1,3,2)
plt.boxplot(scores, vert=False)
plt.title("Boxplot")
```

```python
# Visualization
plt.figure(figsize=(12,4))

# Bar Chart (frequency counts)
plt.subplot(1,3,3)
pd.Series(scores).value_counts().sort_index().plot.bar(color='orange')
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```


---

## üéØ Key Takeaways

* **Mean, Median, Mode** ‚Üí central location.
* **Range, Variance, SD, IQR** ‚Üí spread/variability.
* **Histograms & Boxplots** ‚Üí reveal distribution and outliers.
* **Bar/Pie charts** ‚Üí best for categorical summaries.
